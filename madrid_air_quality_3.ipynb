A new feature is introduced to classify each record into one of three categories—pollutant, weather, or traffic—based on its `magnitude_name`.

```python
# Classify each row by variable type
pollutants = [
    "CO", "NO", "NO2", "NOX", "SO2", "<PM10", "<PM2.5", "O3",
    "TOLUENO", "BENCENO", "ETILBENCENO",
    "HIDROCARBS_TOTALES", "METANO", "HIDROCARBS_NO_METANICOS"
]
weather = ["VV", "DV", "TEMP", "HR", "PRE", "RS", "PRECIPITACION"]
traffic_mask = df["magnitude_name"].str.startswith(("TI_", "SP_", "OC_"), na=False)
traffic_vars = sorted(df.loc[traffic_mask, "magnitude_name"].unique())

def classify_variable(name):
    if name in pollutants:
        return "pollutant"
    elif name in weather:
        return "weather"
    elif name in traffic_vars:
        return "traffic"
    else:
        return "other"

df["variable_type"] = df["magnitude_name"].apply(classify_variable)

# Scale breakdown table
scale_summary = df.groupby("variable_type").size().rename("rows").reset_index()
scale_summary["percentage"] = (scale_summary["rows"] / len(df) * 100).round(2)
display(scale_summary)
```

#### 2.2) Missingness per Variable

```python
miss_per_var = (
    df.groupby(["magnitude_name", "variable_type"])["is_interpolated"]
    .agg(total="count", interpolated="sum")
    .assign(missing_rate=lambda x: (x["interpolated"] / x["total"]).round(3))
    .sort_values("missing_rate", ascending=False)
    .reset_index()
)
display(miss_per_var)
```

#### 2.3) Temporal Gaps — Missing Periods Over Time
A temporal gap occurs when an entire time period has no measurements — not just one sensor, but all sensors for a given variable. We identify these by checking which hours/days have no real (non-interpolated) observations.

```python
# Parse datetime
df["entry_dt"] = pd.to_datetime(df["entry_date"], errors="coerce")
df["date"] = df["entry_dt"].dt.date

# For each variable, count real (non-interpolated) measurements per day
real_counts = (
    df[~df["is_interpolated"]]
    .groupby(["magnitude_name", "date"])
    .size()
    .reset_index(name="real_count")
)

# Days with zero real measurements = temporal gap
all_days = df.groupby(["magnitude_name", "date"]).size().reset_index(name="total_count")
gap_days = all_days.merge(real_counts, on=["magnitude_name", "date"], how="left")
gap_days["real_count"] = gap_days["real_count"].fillna(0)
gap_days["is_gap"] = gap_days["real_count"] == 0

temporal_gaps = (
    gap_days.groupby("magnitude_name")["is_gap"]
    .agg(total_days="count", gap_days="sum")
    .assign(gap_rate=lambda x: (x["gap_days"] / x["total_days"]).round(3))
    .sort_values("gap_days", ascending=False)
    .reset_index()
)
display(temporal_gaps)
```

We visualize temporal gaps for the most affected pollutants as a heatmap — each row is a pollutant, each column is a month, green = data available, red = complete gap.

```python
# Focus on pollutants only
pollutant_gaps = gap_days[gap_days["magnitude_name"].isin(pollutants)].copy()
pollutant_gaps["month"] = pd.to_datetime(pollutant_gaps["date"]).dt.to_period("M").astype(str)

gap_heatmap = (
    pollutant_gaps.groupby(["magnitude_name", "month"])["is_gap"]
    .all()
    .unstack("month")
    .fillna(False)
)

fig, ax = plt.subplots(figsize=(18, 7))
sns.heatmap(
    gap_heatmap.astype(int),
    cmap="RdYlGn_r",
    ax=ax,
    cbar=False,
    linewidths=0
)
ax.set_title("Temporal Gaps per Pollutant (red = complete gap that month)", fontsize=13)
ax.set_xlabel("Month")
ax.set_ylabel("Pollutant")
plt.xticks(rotation=90, fontsize=6)
plt.tight_layout()
plt.show()
```

**What we observe:** Temporal gaps are not random — they tend to cluster in early years when fewer sensors were operational, and some pollutants were not monitored at all until later periods. Gaps that span entire months indicate that a variable was simply not yet introduced into the monitoring network, rather than a sensor malfunction. This reinforces the importance of the time window decision made in Task 1.

#### 2.4) Partial vs Full Timestamp Missingness

The PDF highlights a specific pattern: at a given timestamp, only some variables may be missing while others are present. We distinguish between timestamps where all variables are missing (likely a full sensor outage) versus timestamps where only specific variables are absent (isolated measurement failure).

```python
timestamp_miss = (
    df.groupby(["sensor_name", "entry_date"])["is_interpolated"]
    .agg(total="count", missing="sum")
    .assign(
        all_missing=lambda x: x["missing"] == x["total"],
        partial_missing=lambda x: (x["missing"] > 0) & (x["missing"] < x["total"]),
        none_missing=lambda x: x["missing"] == 0
    )
)

print(f"Timestamps where ALL variables missing  : {timestamp_miss['all_missing'].sum():,}")
print(f"Timestamps where SOME variables missing : {timestamp_miss['partial_missing'].sum():,}")
print(f"Timestamps where NO variables missing   : {timestamp_miss['none_missing'].sum():,}")
```

#### 2.5) Sensor-Specific Gaps
A sensor-specific gap occurs when one station is missing data while others are still reporting. These are more localised and may indicate sensor malfunctions, maintenance periods, or station-level issues.

```python
# For each sensor + variable combination, compute the interpolation rate
sensor_miss = (
    df.groupby(["sensor_name", "magnitude_name"])["is_interpolated"]
    .agg(total="count", interpolated="sum")
    .assign(rate=lambda x: (x["interpolated"] / x["total"]).round(3))
    .reset_index()
    .sort_values("rate", ascending=False)
)

display(sensor_miss[sensor_miss["rate"] > 0.5].head(20))
```

We visualize sensor-specific missingness as a heatmap for pollutants — rows are sensors, columns are pollutants.

```python
# Pivot: sensors × pollutants, value = interpolation rate
sensor_pollutant_miss = (
    df[df["variable_type"] == "pollutant"]
    .groupby(["sensor_name", "magnitude_name"])["is_interpolated"]
    .mean()
    .unstack("magnitude_name")
    .fillna(1.0)  # if a sensor never measured a pollutant, treat as fully missing
)

fig, ax = plt.subplots(figsize=(16, 10))
sns.heatmap(
    sensor_pollutant_miss,
    cmap="Reds",
    ax=ax,
    linewidths=0.3,
    vmin=0,
    vmax=1,
    annot=False
)
ax.set_title("Interpolation Rate per Sensor × Pollutant (darker = more missing)", fontsize=13)
ax.set_xlabel("Pollutant")
ax.set_ylabel("Sensor")
plt.tight_layout()
plt.show()
```

**What we observe:** The heatmap reveals that missingness is highly uneven across sensors. Some stations are almost entirely interpolated for certain pollutants, meaning they contributed almost no real measurements for those variables. These sensor-pollutant combinations should be excluded or treated carefully in any analysis that requires reliable ground-truth data.

#### 2.6) Monthly Interpolation Rate per Sensor Over Time

Beyond knowing which sensors are missing data overall, we want to see *when* they were missing. A sensor might have a long uninterrupted outage in one specific year due to maintenance or malfunction, which would not be visible in an overall average. We plot the monthly interpolation rate per sensor to reveal these temporal patterns.

#### 2.7) Invalid and Inconsistent Values
The is_interpolated flag does not guarantee data quality — real measurements can also be erroneous. We check for several types of invalid values:

```python
issues = {}

# 1. Negative pollutant concentrations (physically impossible)
neg = df[(df["variable_type"] == "pollutant") & (df["value"] < 0)]
issues["Negative pollutant values"] = len(neg)

# 2. Extreme outliers: values beyond 5 standard deviations from the mean per variable
def flag_outliers(group):
    mean = group["value"].mean()
    std  = group["value"].std()
    return group[np.abs(group["value"] - mean) > 5 * std]

outliers = (
    df[df["variable_type"] == "pollutant"]
    .groupby("magnitude_name", group_keys=False)
    .apply(flag_outliers)
)
issues["Extreme outliers (>5σ) in pollutants"] = len(outliers)

# 3. Zero values for variables that should never be zero (e.g. atmospheric pressure)
zero_pressure = df[(df["magnitude_name"] == "PRE") & (df["value"] == 0)]
issues["Zero atmospheric pressure readings"] = len(zero_pressure)

# 4. Duplicate timestamps: same sensor + variable + timestamp with different values
dupes = df.duplicated(subset=["sensor_name", "magnitude_name", "entry_date"], keep=False)
issues["Duplicate sensor/variable/timestamp entries"] = dupes.sum()

for issue, count in issues.items():
    print(f"{issue}: {count:,}")
```

#### 2.8) Chemical Consistency Check — NOX = NO + NO2

Beyond range checks, we can verify internal consistency between chemically related variables. By definition, NOX (nitrogen oxides) is the sum of NO and NO2. Any row where this relationship is badly violated likely represents a sensor calibration error or a data pipeline issue, and is invalid regardless of whether it was flagged as interpolated.

```python
nox_check = df[df["magnitude_name"].isin(["NO", "NO2", "NOX"])].copy()
nox_wide = (
    nox_check.groupby(["sensor_name", "entry_date", "magnitude_name"])["value"]
    .mean()
    .unstack("magnitude_name")
    .dropna()
)

if all(col in nox_wide.columns for col in ["NOX", "NO", "NO2"]):
    nox_wide["expected_NOX"] = nox_wide["NO"] + nox_wide["NO2"]
    nox_wide["nox_error"] = np.abs(nox_wide["NOX"] - nox_wide["expected_NOX"])
    nox_wide["nox_error_pct"] = (
        nox_wide["nox_error"] / nox_wide["expected_NOX"].replace(0, np.nan) * 100
    )

    n_inconsistent = (nox_wide["nox_error_pct"] > 20).sum()
    print(f"Rows where NOX differs from NO+NO2 by more than 20%: {n_inconsistent:,}")
    print(f"Mean absolute error: {nox_wide['nox_error'].mean():.2f}")

    if n_inconsistent > 0:
        display(nox_wide[nox_wide["nox_error_pct"] > 20].head(10))
else:
    print("One or more of NO, NO2, NOX not present in this dataset sample.")
```

## **Task 3 — Imputation**

In this task we reconstruct the original missingness of the dataset and apply **four imputation strategies** to fill the missing values. We then compare all four methods, and METRAQ's own interpolation, to understand which one best performs.

### 3.1) Reconstruct Original Missingness

The METRAQ dataset has already filled originally-missing values via spatial interpolation, flagged by `is_interpolated` = True.

To fairly evaluate our own methods, we must undo this filling: we replace those values with `NaN`, recreating the original gaps. We save the METRAQ-filled values separately to use as a **reference** in the final comparison.


```python
# Working copy — original df is never modified
df_imputation = df.copy()

# Boolean mask: True where a value was originally missing and METRAQ filled it
metraq_mask = df_imputation['is_interpolated'] == True

# Save METRAQ's filled values for later comparison
metraq_values = df_imputation.loc[metraq_mask, 'value'].copy()

# Reconstruct missingness: put NaN back where METRAQ interpolated
df_imputation.loc[metraq_mask, 'value'] = np.nan

print(f"Total records              : {len(df_imputation):,}")
print(f"Originally missing (NaN)   : {metraq_mask.sum():,}  ({metraq_mask.mean()*100:.1f}%)")
print(f"Observed (non-missing)     : {(~metraq_mask).sum():,}  ({(~metraq_mask).mean()*100:.1f}%)")
```

### 3.2) Method 1 — Mean Imputation

The **mean** of all observed readings for a given *(sensor, variable)* group is used to fill missing values.

This is the most basic imputation strategy: it is fast and easy to interpret, but it is sensitive to outliers and ignores any temporal or spatial structure. We use it as a **simple baseline**.

**Fallback:** if an entire *(sensor, variable)* group has no observed values, we fall back to the global mean for that variable across all sensors.

```python
df_mean = df_imputation.copy()

# Fill NaN with the per-(sensor, variable) mean
df_mean['value'] = (
    df_mean
    .groupby(['sensor_name', 'magnitude_name'])['value']
    .transform(lambda x: x.fillna(x.mean()))
)

# Fallback: groups entirely NaN → use global variable mean
global_means = df_imputation.groupby('magnitude_name')['value'].mean()
still_nan = df_mean['value'].isna()
df_mean.loc[still_nan, 'value'] = df_mean.loc[still_nan, 'magnitude_name'].map(global_means)

n_filled = df_mean.loc[metraq_mask, 'value'].notna().sum()
print(f"Mean imputation : filled {n_filled:,} / {metraq_mask.sum():,} originally missing values")
print(f"Still NaN after fallback : {df_mean.loc[metraq_mask, 'value'].isna().sum():,}")
```

### 3.3) Method 2 — Median Imputation

The **median** of all observed readings for a given *(sensor, variable)* group is used to fill missing values.

The median is more robust than the mean because it is not affected by extreme outliers — which are common in air quality data (e.g., pollution spikes during traffic peaks or industrial events). This makes it a better simple baseline than the mean.

**Fallback:** same strategy as mean imputation — global variable median when an entire group is missing.

```python
df_median = df_imputation.copy()

# Fill NaN with the per-(sensor, variable) median
df_median['value'] = (
    df_median
    .groupby(['sensor_name', 'magnitude_name'])['value']
    .transform(lambda x: x.fillna(x.median()))
)

# Fallback: groups entirely NaN → use global variable median
global_medians = df_imputation.groupby('magnitude_name')['value'].median()
still_nan = df_median['value'].isna()
df_median.loc[still_nan, 'value'] = df_median.loc[still_nan, 'magnitude_name'].map(global_medians)

n_filled = df_median.loc[metraq_mask, 'value'].notna().sum()
print(f"Median imputation : filled {n_filled:,} / {metraq_mask.sum():,} originally missing values")
print(f"Still NaN after fallback  : {df_median.loc[metraq_mask, 'value'].isna().sum():,}")
```

### 3.4) Method 3 — KNN Imputation

**K-Nearest Neighbors (KNN)** imputation finds the *K most similar observations* and fills each missing value with their average.

**How we apply it:**
For each variable separately, we reshape the data into a **(timestamp × sensor)** matrix. `sklearn`'s `KNNImputer` then finds, for a timestamp with a missing sensor reading, the K timestamps whose sensor readings are most similar — and uses those to estimate the missing value.

This is a form of **temporal similarity imputation**: a missing reading is estimated from moments when the overall pollution pattern across the city was most similar.

We use `n_neighbors = 5`. If fewer than 5 sensors record a given variable, K is reduced accordingly.

```python
# We work on a copy that shares the same index as df_imputation.
# After imputation we extract values at metraq_mask positions directly —
# no merge needed, no key-alignment issues.
df_knn = df_imputation.copy()

N_NEIGHBORS = 5

for mag in df_imputation['magnitude_name'].unique():

    mag_mask = df_knn['magnitude_name'] == mag

    # Pivot: rows = timestamps, columns = sensors
    pivot = (
        df_knn[mag_mask]
        .pivot_table(index='entry_dt', columns='sensor_name', values='value', aggfunc='mean')
    )

    k = min(N_NEIGHBORS, pivot.shape[1] - 1)

    if k < 1:
        # Only one sensor: fill with its own median
        group_med = df_knn.loc[mag_mask, 'value'].median()
        df_knn.loc[mag_mask, 'value'] = df_knn.loc[mag_mask, 'value'].fillna(group_med)
        continue

    # Fit and transform
    imputer = KNNImputer(n_neighbors=k)
    imputed_array = imputer.fit_transform(pivot)
    pivot_imputed = pd.DataFrame(imputed_array, index=pivot.index, columns=pivot.columns)

    # Stack pivot into a Series with MultiIndex (entry_dt, sensor_name) for fast lookup
    stacked = pivot_imputed.stack()          # shape: (timestamps * sensors,)
    stacked.index.names = ['entry_dt', 'sensor_name']

    # Identify which rows in df_knn need to be updated (originally missing + this variable)
    nan_in_mag = mag_mask & df_knn['value'].isna()

    if nan_in_mag.sum() == 0:
        continue

    # Build lookup keys and retrieve imputed values
    nan_rows = df_knn.loc[nan_in_mag, ['entry_dt', 'sensor_name']]
    keys = list(zip(nan_rows['entry_dt'], nan_rows['sensor_name']))
    imputed_vals = [stacked.get(k, np.nan) for k in keys]
    df_knn.loc[nan_in_mag, 'value'] = imputed_vals

# Global median fallback for any remaining NaN
global_medians_knn = df_imputation.groupby('magnitude_name')['value'].median()
still_nan = df_knn['value'].isna()
df_knn.loc[still_nan, 'value'] = df_knn.loc[still_nan, 'magnitude_name'].map(global_medians_knn)

n_filled = df_knn.loc[metraq_mask, 'value'].notna().sum()
print(f"KNN imputation: filled {n_filled:,} / {metraq_mask.sum():,} originally missing values")
print(f"Still NaN after fallback: {df_knn.loc[metraq_mask, 'value'].isna().sum():,}")
```

### 3.5) Method 4 — Regression Imputation

**Regression imputation** predicts missing values using the other variables available at the same sensor and timestamp as features.

We use `sklearn`'s `IterativeImputer` (also known as **MICE — Multiple Imputation by Chained Equations**). It works as follows:
1. For each variable with missing values, it fits a **regression model** (BayesianRidge by default) using all other variables as predictors.
2. It iterates this process across all variables until the imputed values converge.

**How we apply it:**
For each sensor separately, we reshape the data into a **(timestamp × variable)** matrix and run `IterativeImputer`. This means: if NO2 is missing at sensor X at time T, but NO, NOX, TEMP, and other variables are available at that sensor and time, the model uses those to predict NO2.

This approach is the most **statistically sophisticated** of the four because it leverages cross-variable correlations — which are strong in air quality data (e.g., NO2 ↔ NO ↔ NOX; temperature affects pollutant dispersion).

```python
# Same index-aligned design as KNN: write imputed values directly into df_reg,
# then extract at metraq_mask positions — no merge required.
df_reg = df_imputation.copy()

for sensor in df_imputation['sensor_name'].unique():

    sensor_mask = df_reg['sensor_name'] == sensor

    # Pivot: rows = timestamps, columns = variables
    pivot = (
        df_reg[sensor_mask]
        .pivot_table(index='entry_dt', columns='magnitude_name', values='value', aggfunc='mean')
    )

    if pivot.shape[1] < 2:
        # Only one variable available: fall back to its median
        group_med = df_reg.loc[sensor_mask, 'value'].median()
        df_reg.loc[sensor_mask, 'value'] = df_reg.loc[sensor_mask, 'value'].fillna(group_med)
        continue

    # Iterative regression imputer (MICE), max 10 iterations for speed
    imputer = IterativeImputer(max_iter=10, random_state=42, verbose=0)
    imputed_array = imputer.fit_transform(pivot)
    pivot_imputed = pd.DataFrame(imputed_array, index=pivot.index, columns=pivot.columns)

    # Stack into a Series with MultiIndex (entry_dt, magnitude_name)
    stacked = pivot_imputed.stack()
    stacked.index.names = ['entry_dt', 'magnitude_name']

    # Update only the originally-missing rows for this sensor
    nan_in_sensor = sensor_mask & df_reg['value'].isna()

    if nan_in_sensor.sum() == 0:
        continue

    nan_rows = df_reg.loc[nan_in_sensor, ['entry_dt', 'magnitude_name']]
    keys = list(zip(nan_rows['entry_dt'], nan_rows['magnitude_name']))
    imputed_vals = [stacked.get(k, np.nan) for k in keys]
    df_reg.loc[nan_in_sensor, 'value'] = imputed_vals

# Global median fallback for any remaining NaN
global_medians_reg = df_imputation.groupby('magnitude_name')['value'].median()
still_nan = df_reg['value'].isna()
df_reg.loc[still_nan, 'value'] = df_reg.loc[still_nan, 'magnitude_name'].map(global_medians_reg)

n_filled = df_reg.loc[metraq_mask, 'value'].notna().sum()
print(f"Regression imputation: filled {n_filled:,} / {metraq_mask.sum():,} originally missing values")
print(f"Still NaN after fallback: {df_reg.loc[metraq_mask, 'value'].isna().sum():,}")
```

### 3.6) Build Comparison Table

We now consolidate all four imputed values — plus the METRAQ reference — into a single table, indexed by the originally-missing positions (`is_interpolated == True`).

Each row in this table represents one originally-missing observation, with five value columns:
- `value_metraq` — METRAQ's spatial interpolation (reference)
- `value_mean` — our mean imputation
- `value_median` — our median imputation
- `value_knn` — our KNN imputation
- `value_reg` — our regression imputation

This is the basis for all comparisons below.

```python
# All four imputed dataframes (df_mean, df_median, df_knn, df_reg) share
# the exact same index as df_imputation, so we extract values at metraq_mask
# positions directly — identical to how mean and median are extracted.

missing_rows = (
    df_imputation[metraq_mask][['sensor_name', 'entry_dt', 'magnitude_name', 'variable_type']]
    .copy()
    .reset_index(drop=True)
)

missing_rows['value_metraq'] = metraq_values.values
missing_rows['value_mean']   = df_mean.loc[metraq_mask, 'value'].values
missing_rows['value_median'] = df_median.loc[metraq_mask, 'value'].values
missing_rows['value_knn']    = df_knn.loc[metraq_mask, 'value'].values
missing_rows['value_reg']    = df_reg.loc[metraq_mask, 'value'].values

print(f"Comparison table: {missing_rows.shape[0]:,} rows x {missing_rows.shape[1]} columns")
print("\nNaN count per method:")
for col in ['value_metraq', 'value_mean', 'value_median', 'value_knn', 'value_reg']:
    print(f"  {col:15s}: {missing_rows[col].isna().sum():,} NaN")

missing_rows.sample(5)
```

```python
# count rows where KNN and Regression differ in the comparison table
diff_count = (missing_rows["value_knn"] != missing_rows["value_reg"]).sum()
print(f"Number of records where value_knn != value_reg: {diff_count}")
```

### 3.7) Distribution Comparison — KDE Plots

Since we have no ground truth, we evaluate imputation quality by asking:

> *Does the imputed distribution resemble METRAQ's reference distribution?*

A good imputation method should produce a value distribution that is statistically consistent with the METRAQ reference — it should not over-concentrate values (as mean/median tend to do) nor produce implausible extremes.

We use **Kernel Density Estimation (KDE)** plots to visually compare the distribution of imputed values across all five methods, broken down by variable type.

```python
METHOD_STYLES = {
    'METRAQ (ref)' : ('steelblue',  2.5, '--'),
    'Mean'         : ('tomato',     1.8, '-'),
    'Median'       : ('darkorange', 1.8, '-'),
    'KNN'          : ('green',      1.8, '-'),
    'Regression'   : ('purple',     1.8, '-')
}

METHOD_COLS = {
    'METRAQ (ref)' : 'value_metraq',
    'Mean'         : 'value_mean',
    'Median'       : 'value_median',
    'KNN'          : 'value_knn',
    'Regression'   : 'value_reg'
}

var_types = sorted(missing_rows['variable_type'].dropna().unique())
fig, axes = plt.subplots(1, len(var_types), figsize=(7 * len(var_types), 5))
if len(var_types) == 1:
    axes = [axes]

for ax, vtype in zip(axes, var_types):
    sub = missing_rows[missing_rows['variable_type'] == vtype]

    for label, col in METHOD_COLS.items():
        vals = sub[col].dropna()
        if len(vals) < 10:
            continue
        lo, hi = vals.quantile(0.01), vals.quantile(0.99)
        if lo == hi:
            continue
        vals_clipped = vals.clip(lo, hi)
        color, lw, ls = METHOD_STYLES[label]
        kde = gaussian_kde(vals_clipped)
        x = np.linspace(lo, hi, 300)
        ax.plot(x, kde(x), label=label, color=color, linewidth=lw, linestyle=ls)

    ax.set_title(f'{vtype.capitalize()}', fontsize=13, fontweight='bold')
    ax.set_xlabel('Imputed value')
    ax.set_ylabel('Density')
    ax.legend(fontsize=9)

plt.suptitle('KDE — Distribution of Imputed Values by Variable Type', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

### 3.8) Quantitative Summary — MAD and KS Test

We use two quantitative metrics to compare the four methods against METRAQ:

- **MAD (Mean Absolute Difference):** average absolute difference between each method's imputed values and METRAQ's values at the same positions. Lower is better.
- **KS statistic (Kolmogorov–Smirnov test):** measures how different two distributions are (0 = identical, 1 = completely different). Lower means a more similar distribution to METRAQ.

These metrics tell us **which method most closely reproduces METRAQ's behavior**

```python
method_cols = {
    'Mean'       : 'value_mean',
    'Median'     : 'value_median',
    'KNN'        : 'value_knn',
    'Regression' : 'value_reg',
}

summary_rows = []

for vtype in sorted(missing_rows['variable_type'].dropna().unique()):
    sub = missing_rows[missing_rows['variable_type'] == vtype].dropna(
        subset=['value_metraq'] + list(method_cols.values())
    )
    if sub.empty:
        continue

    metraq_vals = sub['value_metraq'].values

    for method, col in method_cols.items():
        method_vals = sub[col].values
        mad   = np.abs(method_vals - metraq_vals).mean()
        ks_stat, ks_pval = ks_2samp(metraq_vals, method_vals)
        summary_rows.append({
            'Variable type' : vtype,
            'Method'        : method,
            'Mean'          : method_vals.mean(),
            'Std'           : method_vals.std(),
            'MAD vs METRAQ' : mad,
            'KS statistic'  : ks_stat,
            'KS p-value'    : ks_pval,
        })

summary_df = (
    pd.DataFrame(summary_rows)
    .set_index(['Variable type', 'Method'])
    .round(3)
)

display(summary_df)
```

##### Analysis by Variable Type

Pollutant Variables:

- **Median** is well performing: it achieves the **lowest KS statistic (0.122)** and the **lowest MAD vs METRAQ (15.77)**, meaning it both preserves the original distribution best and stays closest to the reference imputation (robust to outliers).
- **Mean** performs moderately but inflates the distribution due to sensitivity to outliers, pollutant data is typically right-skewed, making the mean a poor central estimate.
- **KNN** performs similarly to Mean, with a slightly higher KS statistic (0.171), suggesting it is slightly worse at recovering the original distribution shape.
- **Regression** has failed: it produces a mean of $-2.3 \times 10^{11}$ and a standard deviation of $1.3 \times 10^{13}$, which are entirely unphysical --> **numerical instability**. This is likely due to multicollinearity or poorly conditioned predictors. Therefore we can clearly disqualify this method for pollutant imputation.

Weather Variables:

- **Median and KNN are tied** on the KS statistic (0.151), both substantially outperforming Mean (0.190).
- KNN achieves a marginally lower MAD vs METRAQ (49.07 vs 49.08), giving it a slight edge as the most principled method.
- **Mean** is not well performing, and has the highest KS statistic (0.190) and the highest MAD (56.64).
- **Regression** again produces unstable values (mean $\approx -1.5 \times 10^8$, std $\approx 1.5 \times 10^{10}$), despite coincidentally achieving the same KS score (0.151) as Median and KNN. This is likely a coincidence of distribution shape, not a sign of correctness, therefore the individual imputed values are wildly wrong.




> **Median imputation is the recommended method overall.**

It is the most robust strategy given the nature of the data:
- Pollutant and weather measurements are often **skewed and contain outliers**, making median-based estimates more reliable than mean-based ones.
- It achieves the **best or tied-best performance** on both KS statistic and MAD vs METRAQ across both variable types.
- It is simple, interpretable, and computationally trivial.
